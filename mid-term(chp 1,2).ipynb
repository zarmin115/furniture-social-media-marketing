```python
#palimdrome
num = int(input("Enter number: "))
temp = num
rev = 0

while num > 0:
    digit = num % 10
    rev = rev * 10 + digit
    num = num // 10

if temp == rev:
    print("Palindrome Number")
else:
    print("Not Palindrome")
```

```python
#list for add,display,search,delete function
students = []

def add_student():
    id = input("Enter ID: ")
    name = input("Enter Name: ")
    roll = input("Enter Roll No: ")

    data = [id, name, roll]
    students.append(data)

    print("Student Added")

def display_student():
    for s in students:
        print("ID:", s[0], "Name:", s[1], "Roll:", s[2])

def search_student():
    id = input("Enter ID to search: ")

    for s in students:
        if s[0] == id:
            print("Student Found:", s)
            return

    print("Student Not Found")

def update_student():
    id = input("Enter ID to update: ")

    for s in students:
        if s[0] == id:
            name = input("Enter new name: ")
            roll = input("Enter new roll no: ")

            s[1] = name
            s[2] = roll

            print("Student Updated")
            return

    print("Student Not Found")

def delete_student():
    id = input("Enter ID to delete: ")

    for s in students:
        if s[0] == id:
            students.remove(s)
            print("Student Deleted")
            return

    print("Student Not Found")

while True:

    print("\n--- MENU ---")
    print("1 Add Student")
    print("2 Display Student")
    print("3 Search Student")
    print("4 Update Student")
    print("5 Delete Student")
    print("6 Exit")

    ch = int(input("Enter choice: "))

    if ch == 1:
        add_student()

    elif ch == 2:
        display_student()

    elif ch == 3:
        search_student()

    elif ch == 4:
        update_student()

    elif ch == 5:
        delete_student()

    elif ch == 6:
        break

    else:
        print("Invalid Choice")
```

```python
#dictionary forr add,display,search delete

students = {}

def add_student():
    id = input("Enter ID: ")
    name = input("Enter Name: ")
    roll = input("Enter Roll No: ")

    students[id] = {"name": name, "roll": roll}
    print("Student Added")

def display_student():
    for k,v in students.items():
        print("ID:", k, "Name:", v["name"], "Roll:", v["roll"])

def search_student():
    id = input("Enter ID to search: ")
    if id in students:
        print("Student Found:", students[id])
    else:
        print("Student Not Found")

def update_student():
    id = input("Enter ID to update: ")

    if id in students:
        name = input("Enter new name: ")
        roll = input("Enter new roll no: ")

        students[id]["name"] = name
        students[id]["roll"] = roll

        print("Student Updated")
    else:
        print("Student Not Found")

def delete_student():
    id = input("Enter ID to delete: ")
    if id in students:
        students.pop(id)
        print("Student Deleted")
    else:
        print("Student Not Found")

while True:

    print("\n--- MENU ---")
    print("1 Add Student")
    print("2 Display Student")
    print("3 Search Student")
    print("4 Update Student")
    print("5 Delete Student")
    print("6 Exit")

    ch = int(input("Enter choice: "))

    if ch == 1:
        add_student()

    elif ch == 2:
        display_student()

    elif ch == 3:
        search_student()

    elif ch == 4:
        update_student()

    elif ch == 5:
        delete_student()

    elif ch == 6:
        break

    else:
        print("Invalid Choice")
```
