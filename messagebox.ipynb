```python
import tkinter as tk
from tkinter import messagebox
root = tk.Tk()
root.configure(bg="yellow")
root.geometry("700x700")
root.title("registrastion from")

def show():
    username=name.get()
    password=pwd.get()
    messagebox.showinfo("registrastion form details",f"username={username},password={password}")
    
username_lb = tk.Label(root,text="enter username",font=("Arial",12),bg="lightblue")
username_lb.grid(row=0,column=0,padx=10,pady=10,sticky="nswe")
name = tk.Entry(root,font=("Arial",12))
name.grid(row=0,column=1,padx=10,pady=10,sticky="nsew")

password_lb = tk.Label(root,text="enter password",font=("Arial",12),bg="lightblue")
password_lb.grid(row=1,column=0,padx=10,pady=10,sticky="nswe")
pwd = tk.Entry(root,font=("Arial",12))
pwd.grid(row=1,column=1,padx=10,pady=10,sticky="nsew")

gender = tk.Label(root,text="gender",font=("Arial",12),bg="lightblue")
gender.grid(row=0,column=0,padx=10,pady=10,sticky="nswe")

btn=tk.Button(root,text="show",activebackground="yellow",activeforeground="red",
              bg="lightblue",font=("Helvetica",14),command=show)
btn.grid(row=2,column=1,padx=10,pady=10,sticky="w")


root.mainloop()
```
